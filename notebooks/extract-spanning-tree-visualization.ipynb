```python
from graphtheory.edge_optimize import *
import pyzx as zx
import random
```

```python
N = 10
random_graph = nx.erdos_renyi_graph(N,0.3)
g = Graph()
vertex_dict = {v: g.add_vertex(VertexType.Z,random.random()*N/2,random.random()*N/2) for v in list(random_graph.nodes) if random_graph.degree(v) > 0}
for u,v, _ in random_graph.edges.data():
    g.add_edge(g.edge(vertex_dict[u],vertex_dict[v]), EdgeType.HADAMARD)
g.normalize()
zx.draw(g, labels=True, scale=50)
```

```python
independent_set = find_maximal_independent_set(g)
print(independent_set)
lcomp_edge_optimize_greedy(g, independent_set)
zx.draw(g, labels=True)
vertex_degree_function = [(v,len(g.neighbors(v))) for v in g.vertex_set()]
start_vertex, degree = min(vertex_degree_function, key = lambda x: x[1])
# start_vertex = list(g.vertices())[0]
print("choose start vertex",start_vertex, degree)
tree = extract_spanning_tree_dfs(g, start_vertex)
zx.draw(tree, labels=True)
```

```python
tree_dict = dict()
import json
for v in tree.vertices():
    tree_dict[v] = list(tree.neighbors(v))
with open('tree.json', 'w') as f:
    json.dump(tree_dict, f)
```
