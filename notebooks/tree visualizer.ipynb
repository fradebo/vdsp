```python
#Visualize trees in dict format (similar to json format as required by first passage time calculation)
import networkx as nx
```

```python
gdict = {7: [1, 4, 5, 9, 10], 10: [3, 0, 7], 3: [10], 0: [8, 10], 8: [6, 0], 6: [8], 9: [2, 7], 2: [9], 5: [7], 4: [7], 1: [7]}
g = nx.Graph(gdict)
nx.draw(g,with_labels=True)
```
